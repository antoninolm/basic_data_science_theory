## Bar Chart with Error

First, we are going to look at the chart called bar_chart.png.



The data you will need to recreate this chart is in the lists past_years_averages, years, and error.
Save your recreated chart to a file called my_bar_chart.png.

```python
from matplotlib import pyplot as plt

past_years_averages = [82, 84, 83, 86, 74, 84, 90]
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006]
error = [1.5, 2.1, 1.2, 3.2, 2.3, 1.7, 2.4]

# Checkpoint 1: First, create a figure of width 10 and height 8.
plt.figure(figsize=(10, 8))

# Checkpoint 2: Plot the blue bars, which have the heights listed in past_years_averages.
# Checkpoint 3: Add error bars of cap size 5 and heights corresponding to the list error.
plt.bar(range(len(past_years_averages)), past_years_averages, yerr=error, capsize=5)

# Checkpoint 4: Set the axis to go from -0.5 to 6.5 on the x-axis and 70 to 95 on the y-axis.
plt.axis([-0.5, 6.5, 70, 95])

# Checkpoint 5: Create an ax object using plt.subplot().
# Use ax to set the x-axis ticks to be range(len(years)) and the x-axis labels to be the years list.
ax = plt.subplot()
ax.set_xticks(range(len(years)))
ax.set_xticklabels(years)

# Checkpoint 6: Add the title "Final Exam Averages", x-axis label "Year", and y-axis label "Test average".
plt.title("Final Exam Averages")
plt.xlabel("Year")
plt.ylabel("Test average")

# Display the chart
plt.show()

# Checkpoint 7: Save your figure to a file called my_bar_chart.png.
plt.savefig("saved_images/my_bar_chart.png")
```

## Side By Side Bars
Now, we are going to look at the chart called side-by-side.png. This displays the differences in average test scores between students who went to two different middle schools before enrolling in your high school.



The data you will need to recreate this chart is in the lists middle_school_a, middle_school_b, and unit_topics.
Save your recreated chart to a file called my_side_by_side.png.

```python
from matplotlib import pyplot as plt

unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
middle_school_a = [80, 85, 84, 83, 86]
middle_school_b = [73, 78, 77, 82, 86]

def create_x(t, w, n, d):
    return [t * x + w * n for x in range(d)]

# Checkpoint 1: Using create_x, make the lists school_a_x and school_b_x
# which will determine where to put the bars for Middle School A and Middle School B.
school_a_x = create_x(2, 0.8, 1, len(middle_school_a))
school_b_x = create_x(2, 0.8, 2, len(middle_school_b))

# Checkpoint 2: Create a figure of width 10 and height 8.
plt.figure(figsize=(10, 8))

# Checkpoint 3: Create a set of axes and save them to ax.
ax = plt.subplot()

# Checkpoint 4: Plot a set of bars representing middle_school_a
# and a set representing middle_school_b next to each other on the same graph.
plt.bar(school_a_x, middle_school_a, label="Middle School A")
plt.bar(school_b_x, middle_school_b, label="Middle School B")

# Checkpoint 5: Create a new list of x-values called middle_x,
# which are the values in the middle of school_a_x and school_b_x.
middle_x = [(a + b) / 2.0 for a, b in zip(school_a_x, school_b_x)]

# Checkpoint 6: Set the x-ticks to be the middle_x list.
ax.set_xticks(middle_x)

# Checkpoint 7: Set the x-tick labels to be the list unit_topics.
ax.set_xticklabels(unit_topics)

# Checkpoint 8: Create a legend that labels the first set of bars "Middle School A"
# and the second set "Middle School B".
plt.legend()

# Checkpoint 9: Create a title, x-axis label, and y-axis label.
plt.title("Test Averages on Different Units")
plt.xlabel("Unit")
plt.ylabel("Test Average")

# Display the chart
plt.show()

# Checkpoint 10: Save your figure to a file called my_side_by_side.png.
plt.savefig("saved_images/my_side_by_side.png")
```

## Stacked Bar
Now, we are going to look at the chart called stacked-bars.png. This graph displays the breakdown of students who got As, Bs, Cs, Ds, and Fs in each unit.



The data you will need to recreate this chart is in the lists As, Bs, Cs, Ds, Fs, and unit_topics. Save your recreated chart to a file called my_stacked_bar.png.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
As = [6, 3, 4, 3, 5]
Bs = [8, 12, 8, 9, 10]
Cs = [13, 12, 15, 13, 14]
Ds = [2, 3, 3, 2, 1]
Fs = [1, 0, 0, 3, 0]

x = range(5)

# Checkpoint 1: The bottom of the bars representing the Cs will be at the height of the As plus the Bs.
c_bottom = np.add(As, Bs)

# Define d_bottom (where the Cs end) and f_bottom (where the Ds end).
d_bottom = np.add(c_bottom, Cs)
f_bottom = np.add(d_bottom, Ds)

# Checkpoint 2: Create a figure of width 10 and height 8.
plt.figure(figsize=(10, 8))

# Checkpoint 3: Plot the As, Bs, Cs, Ds, and Fs.
# Give each one the appropriate bottom list so they stack on top of each other.
plt.bar(range(len(unit_topics)), As)
plt.bar(range(len(unit_topics)), Bs, bottom=As)
plt.bar(range(len(unit_topics)), Cs, bottom=c_bottom)
plt.bar(range(len(unit_topics)), Ds, bottom=d_bottom)
plt.bar(range(len(unit_topics)), Fs, bottom=f_bottom)

# Checkpoint 4: Create a set of axes and save them to ax.
ax = plt.subplot()

# Checkpoint 5: Set the x-ticks to be range(len(unit_topics)).
ax.set_xticks(range(len(unit_topics)))

# Checkpoint 6: Set the x-tick labels to be the unit_topics.
ax.set_xticklabels(unit_topics)

# Checkpoint 7: Give the plot the title, x-axis label, and y-axis label.
plt.title('Grade distribution')
plt.xlabel('Unit')
plt.ylabel('Number of Students')

# Display the chart
plt.show()

# Checkpoint 8: Save your figure to a file called my_stacked_bar.png.
plt.savefig('saved_images/my_stacked_bar.png')
```

## Labeled Pie Chart
Now, we are going to look at the chart called pie.png. This displays what students think the hardest topic covered throughout your math course is.



The data you will need to recreate this chart is in the lists unit_topics and num_hardest_reported.
Save your recreated chart to a file called my_pie_chart.png.

```python
from matplotlib import pyplot as plt

unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
num_hardest_reported = [1, 3, 10, 15, 1]

# Checkpoint 1: Create a figure of width 10 and height 8.
plt.figure(figsize=(10, 8))

# Checkpoint 2: Plot the num_hardest_reported list as a pie chart.
# Checkpoint 4: Label the slices with the unit_topics list and put a percentage label
# on each slice, rounded to the nearest int.
plt.pie(num_hardest_reported, labels=unit_topics, autopct='%1d%%')

# Checkpoint 3: Set the axes to be 'equal'.
plt.axis("equal")

# Checkpoint 5: Add the title "Hardest Topics".
plt.title("Hardest Topics")

# Display the chart
plt.show()

# Checkpoint 6: Save your figure to a file called my_pie_chart.png.
plt.savefig('saved_images/my_pie_chart.png')
```

## Line with Shaded Error
Now, we are going to look at the chart called line_graph.png. This displays the relationship between the time students say they studied and the scores they received on their final exams.



The data you will need to recreate this chart is in the lists hours_reported and exam_scores.
Save your recreated chart to a file called my_line_graph.png.

```python
from matplotlib import pyplot as plt

hours_reported = [3, 2.5, 2.75, 2.5, 2.75, 3.0, 3.5, 3.25, 3.25, 3.5, 
                  3.5, 3.75, 3.75, 4, 4.0, 3.75, 4.0, 4.25, 4.25, 4.5, 
                  4.5, 5.0, 5.25, 5, 5.25, 5.5, 5.5, 5.75, 5.25, 4.75]

exam_scores = [52.53, 59.05, 61.15, 61.72, 62.58, 62.98, 64.99, 67.63, 68.52, 70.29, 
               71.33, 72.15, 72.67, 73.85, 74.44, 75.62, 76.81, 77.82, 78.16, 78.94, 
               79.08, 80.31, 80.77, 81.37, 85.13, 85.38, 89.34, 90.75, 97.24, 98.31]

# Checkpoint 1: Create a figure of width 10 and height 8.
plt.figure(figsize=(10, 8))

# Checkpoint 3: Let’s assume the error on students’ reporting of their hours studying is 20%.
# Create a list hours_lower_bound (20% less) and hours_upper_bound (20% more).
hours_lower_bound = [element - (element * 0.2) for element in hours_reported]
hours_upper_bound = [element + (element * 0.2) for element in hours_reported]

# Checkpoint 2: Plot the exam_scores on the x-axis and the hours_reported on the y-axis,
# with a linewidth of 2.
plt.plot(exam_scores, hours_reported, linewidth=2)

# Checkpoint 4: Shade the area between hours_lower_bound and hours_upper_bound
# with an alpha of 0.2 for transparency.
plt.fill_between(exam_scores, hours_lower_bound, hours_upper_bound, alpha=0.2, color="skyblue")

# Checkpoint 5: Add title and axis labels.
plt.title("Time spent studying vs final exam scores")
plt.xlabel('Score')
plt.ylabel('Hours studying (self-reported)')

# Display the chart
plt.show()

# Checkpoint 6: Save your figure to a file called my_line_graph.png.
plt.savefig('saved_images/my_line_graph.png')
```
