## Basic Line Plot
Line graphs are particularly useful for illustrating how a variable evolves over time or across a continuous range. They can be used to represent information such as gasoline prices across a decade, a person’s changing weight over several months, or temperature variations along a line of longitude. In Matplotlib, a simple line graph can be created using the `.plot()` method, which draws the line based on pairs of x- and y-values, and displayed with `.show()`. For example, if you define `x_values` as a list of x-coordinates and `y_values` as the corresponding y-coordinates, calling `plt.plot(x_values, y_values)` generates the line, while `plt.show()` renders it visibly on the screen. This forms the basis of building visualizations: the x-values determine positions along the horizontal axis, the y-values determine the vertical axis, and Matplotlib handles connecting these pairs into a coherent line. With these simple commands, you can quickly begin to translate lists of numbers into visual insights, forming the foundation for more advanced plotting techniques.


```python
# Define the x-axis values: days of the week, represented numerically (0 through 6)
days = [0, 1, 2, 3, 4, 5, 6]

# Define the y-axis values: money spent on lunch corresponding to each day
money_spent = [10, 12, 12, 10, 14, 22, 24]

# Plot the data: 'days' on the x-axis, 'money_spent' on the y-axis
plt.plot(days, money_spent)

# Display the line graph in a window or output cell
plt.show()
```

## Basic Line Plot II - Multiple lines
A line graph can contain more than one dataset, which is useful when comparing two variables that share the same scale and categories. In Matplotlib, this is achieved simply by calling `plt.plot()` multiple times before displaying the figure. Each call adds a new line to the same set of axes, and Matplotlib automatically assigns different default colors to distinguish them. For example, if you already have a graph showing your lunch spending across the week, you can add a friend’s spending by plotting their values with the same list of days on the x-axis. The first dataset will appear as a blue line and the second as an orange line. This ability to overlay multiple plots makes it straightforward to visualize comparisons and identify similarities or differences between two related patterns.


```python
# Define the x-axis values: time steps (e.g., days or weeks)
time = [0, 1, 2, 3, 4]

# Define the y-axis values for revenue at each point in time
revenue = [200, 400, 650, 800, 850]

# Define the y-axis values for costs at each point in time
costs = [150, 500, 550, 550, 560]

# Plot revenue against time (this will be the first line, default color blue)
plt.plot(time, revenue)

# Plot costs against time on the same axes (second line, default color orange)
plt.plot(time, costs)

# Display the figure with both lines
plt.show()
```

## Linestyles
When plotting multiple lines on the same graph, customizing their appearance helps distinguish the datasets more clearly. In Matplotlib, you can change the color of a line using the `color` keyword, which accepts either HTML color names like `'green'` or HEX codes such as `'#AAAAAA'`. Beyond color, the style of the line can also be adjusted with the `linestyle` keyword. Common options include `'--'` for dashed lines, `':'` for dotted lines, or even an empty string to remove the connecting line entirely. To emphasize individual data points, you can add markers with the `marker` keyword, choosing from shapes like `'o'` for circles, `'s'` for squares, or `'*'` for stars. These customizations can be combined in a single plot call, allowing you to adjust color, line style, and marker at once. For example, a green dashed line could represent one dataset, while another is displayed in gray with circular markers. This flexibility ensures that graphs are not only accurate but also visually clear and easy to interpret.

Here’s a quick set of examples with 5 different values each that you can use when customizing plots in Matplotlib:

- Colors (you can use names or HEX codes):
'red', 'blue', 'orange', 'purple', '#00FF00' (bright green in HEX)
- Markers (shapes for data points):
'x' (x-shaped marker), 'D' (diamond), '^' (upward triangle), 'v' (downward triangle), 'p' (pentagon)
- Linestyles (style of the connecting line):
'-' (solid line, default), '--' (dashed), ':' (dotted), '-.' (dash-dot), '' (no line, only markers are shown)

```python
# Define the x-axis values: time intervals (e.g., 0 through 4)
time = [0, 1, 2, 3, 4]

# Define the y-axis values: revenue values corresponding to each time
revenue = [200, 400, 650, 800, 850]

# Define the y-axis values: costs values corresponding to each time
costs = [150, 500, 550, 550, 560]

# Plot revenue vs time with a purple dashed line
plt.plot(time, revenue, color="purple", linestyle="--")

# Plot costs vs time with a minty green HEX color and square markers
plt.plot(time, costs, color="#82edc9", marker="s")

# Display the figure with both customized lines
plt.show()
```

## Axis
When working with plots, it is often useful to adjust the visible portion of the graph to focus on specific details. Matplotlib allows this through the `plt.axis()` function, which sets the ranges for both axes. You provide it with a list of four values: the minimum and maximum for the x-axis, followed by the minimum and maximum for the y-axis. For instance, calling `plt.axis([0, 3, 2, 5])` restricts the view so that only the section of the graph where x ranges from 0 to 3 and y ranges from 2 to 5 is shown. This creates a zoomed-in effect that highlights a particular area of interest. The feature is particularly useful when the overall dataset spans a wide range, but you want to closely examine local variations or patterns within a smaller window. By combining `plt.axis()` with your plots, you can quickly shift perspective and tailor the visualization to emphasize the most relevant part of your data.

## Subplots
Sometimes a single set of axes is not enough to convey the relationships you want to highlight, and placing multiple plots side by side can make comparisons clearer. In Matplotlib, each separate set of axes within a figure is called a subplot, while the overall container that holds them is known as the figure. Subplots can be arranged in grids of rows and columns, and the function `plt.subplot()` is used to define where each one belongs. This function takes three arguments: the number of rows, the number of columns, and the index of the current subplot within that grid. For example, `plt.subplot(2, 3, 4)` creates the fourth subplot in a layout with two rows and three columns. Any plotting command that follows will be applied to that specific subplot. This allows you to build figures with multiple plots that each have their own axes, titles, and data. For instance, you might create two subplots side by side, one showing a green line and the other a blue one, each with its own title. By using subplots, you can present multiple perspectives or datasets within the same figure, making your analysis both compact and visually organized.

```python
# Define the x-axis values: months of the year (0 through 11)
months = range(12)

# Define the y-axis values: average monthly temperature
temperature = [36, 36, 39, 52, 61, 72, 77, 75, 68, 57, 48, 48]

# Define the y-axis values: number of flights to Hawaii each month
flights_to_hawaii = [1200, 1300, 1100, 1450, 850, 750, 400, 450, 400, 860, 990, 1000]

# Create the first subplot: 1 row, 2 columns, position 1 (left)
# Plot temperature vs. months as a line graph
plt.subplot(1, 2, 1)
plt.plot(months, temperature)

# Create the second subplot: 1 row, 2 columns, position 2 (right)
# Plot flights to Hawaii vs. temperature as a scatter plot (using 'o' markers)
plt.subplot(1, 2, 2)
plt.plot(temperature, flights_to_hawaii, marker="o")

# Display both subplots side by side
plt.show()
```

## Subplot pt. 2
When creating figures with multiple subplots, the default layout can sometimes cause overlapping elements such as axis labels, tick marks, or titles, which reduces readability. Matplotlib provides the `plt.subplots_adjust()` function to fine-tune the spacing and margins within a figure. This function accepts keyword arguments that control the margins on each side (`left`, `right`, `bottom`, `top`) as well as the spacing between subplots (`wspace` for horizontal space and `hspace` for vertical space). For example, if a y-axis label is being cut off, increasing the `left` margin provides extra room, while adjusting `bottom` allows space for x-axis labels. Similarly, reducing overlap between subplots arranged in rows or columns can be achieved by modifying `wspace` or `hspace`. These parameters can be used individually or in combination—for instance, `plt.subplots_adjust(top=0.95, hspace=0.25)` both increases the top margin and enlarges the vertical space between plots. By applying such adjustments, you can make sure your multi-plot figures remain clear, balanced, and easy to interpret.

```python
x = range(7)
straight_line = [0, 1, 2, 3, 4, 5, 6]
parabola = [0, 1, 4, 9, 16, 25, 36]
cubic = [0, 1, 8, 27, 64, 125, 216]

# Top subplot: 2 rows, 1 column, position 1
# - "2" = total number of rows in the figure
# - "1" = total number of columns in the figure
# - "1" = index of the subplot to use (top row, only column)
plt.subplot(2, 1, 1)
plt.plot(x, straight_line)

# Bottom-left subplot: 2 rows, 2 columns, position 3
# - "2" = total number of rows in the figure
# - "2" = total number of columns in the figure
# - "3" = index of the subplot (count goes left to right, top to bottom)
#          so 3 means 2nd row, 1st column
plt.subplot(2, 2, 3)
plt.plot(x, parabola)

# Bottom-right subplot: 2 rows, 2 columns, position 4
# - "2" = total number of rows in the figure
# - "2" = total number of columns in the figure
# - "4" = index of the subplot (2nd row, 2nd column)
plt.subplot(2, 2, 4)
plt.plot(x, cubic)

# Adjust subplot layout:
# - wspace=0.35 adds horizontal space between subplots in the bottom row
# - bottom=0.2 increases the bottom margin to make space for labels/ticks
plt.subplots_adjust(wspace=0.35, bottom=0.2)

# Display the figure with all three subplots arranged correctly
plt.show()
```

## Legends
When plotting multiple lines on the same graph, it is important to make clear which line corresponds to which dataset. This is done by adding a legend with `plt.legend()`. The legend can take a list of labels that describe each line, and Matplotlib will match them in the order that the lines were plotted. For example, if you plot one curve for a parabola and another for a cubic, you can call `plt.legend(['parabola', 'cubic'])` to display a small box identifying each line. By default, the legend will appear in the “best” position, which Matplotlib determines automatically, but you can override this using the `loc` keyword. Accepted values for `loc` include both descriptive strings such as `'upper right'` or `'center'`, and numeric codes from 0 to 10, with 0 meaning “best.” For instance, setting `loc=6` will place the legend along the left center of the figure. Another approach is to assign labels directly inside each `plt.plot()` call using the `label` keyword, and then simply call `plt.legend()` without arguments. This can make your code cleaner and easier to follow. In either case, using legends ensures that viewers can quickly identify and distinguish between multiple datasets on the same graph.

```python
# Define the x-axis values: months (0 through 11)
months = range(12)

# Define the y-axis values: temperatures in three locations
hyrule = [63, 65, 68, 70, 72, 72, 73, 74, 71, 70, 68, 64]
kakariko = [52, 52, 53, 68, 73, 74, 74, 76, 71, 62, 58, 54]
gerudo = [98, 99, 99, 100, 99, 100, 98, 101, 101, 97, 98, 99]

# Plot the three datasets as separate lines
plt.plot(months, hyrule)
plt.plot(months, kakariko)
plt.plot(months, gerudo)

# Create a list of legend labels for each dataset
legend_labels = ["Hyrule", "Kakariko", "Gerudo Valley"]

# Add the legend to the chart, positioned at the lower center
plt.legend(legend_labels, loc="lower center")

# Display the final chart
plt.show()
```

## Modify Ticks
When customizing a plot, you often want more control over the tick marks along the axes than Matplotlib’s defaults provide. To do this, you work with an **axes object**, which is created by calling `ax = plt.subplot()`. Even if you only have one plot, assigning it to an axes object gives you access to methods for modifying ticks and labels.
You can change where ticks appear using `ax.set_xticks()` for the x-axis or `ax.set_yticks()` for the y-axis, and pass in a list of positions. By default, the labels will match those numeric values, but you can replace them with custom strings using `ax.set_xticklabels()` or `ax.set_yticklabels()`. For instance, if you want the y-axis ticks at `0.1`, `0.6`, and `0.8` but displayed as `10%`, `60%`, and `80%`, you set the tick positions with `ax.set_yticks([0.1, 0.6, 0.8])` and then map them to text labels with `ax.set_yticklabels(['10%', '60%', '80%'])`.
This flexibility is particularly useful when your raw data values are not the most intuitive way to present information. For example, you may want to use years instead of index numbers on the x-axis, or percentages instead of decimals on the y-axis. By working with the axes object, you gain fine-grained control to make your plots more readable and tailored to your audience.

```python
# Define the labels for the x-axis: month abbreviations
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", 
               "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Define the x-axis values: months as numbers 0–11
months = range(12)

# Define the y-axis values: proportion of users converted each month
conversion = [0.05, 0.08, 0.18, 0.28, 0.4, 0.66, 
              0.74, 0.78, 0.8, 0.81, 0.85, 0.85]

# Add axis labels
plt.xlabel("Months")
plt.ylabel("Conversion")

# Plot the conversion line
plt.plot(months, conversion)

# Create an axes object to control ticks and labels
ax = plt.subplot()

# Set the x-ticks to match the numeric months
ax.set_xticks(months)

# Replace the x-tick labels with the month abbreviations
ax.set_xticklabels(month_names)

# Set the y-tick positions at meaningful decimal values
ax.set_yticks([0.10, 0.25, 0.5, 0.75])

# Replace the y-tick labels with percentage strings
ax.set_yticklabels(["10%", "25%", "50%", "75%"])

# Display the customized plot
plt.show()
```

# Beyond the Line Chart

## Simple Bar Chart
A bar chart compares values across categories, with bar heights representing the data. In Matplotlib, you create one using `plt.bar(x_values, y_values)`. The x-values are usually a simple sequence like `range(len(data))`, while the y-values are your measurements. For example, plotting `days_in_year` with `range(len(days_in_year))` gives a basic bar chart. Although the plot shows comparisons clearly, labels are needed to make it fully understandable.


```python
# Categories of drinks sold
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]

# Number of drinks sold in the past month (aligned with the 'drinks' list)
sales = [91, 76, 56, 66, 52, 27]

# Plot a bar chart:
# - x-values: range(len(drinks)) → [0, 1, 2, 3, 4, 5]
# - y-values: sales numbers
plt.bar(range(len(drinks)), sales)

# Display the bar chart
plt.show()
```

## Simple bar Chart II
When making bar charts, it’s important that each bar clearly corresponds to its category. By default, bars are positioned at numeric values, but without labels, it’s hard to interpret what each bar means. To fix this, we first create an axes object with `ax = plt.subplot()`. Then, we set the x-tick positions to match the bar positions (for example, `[0, 1, 2, ...]`). Finally, we assign meaningful labels with `ax.set_xticklabels()`, using the category names, such as drink types. If labels are long or overlap, we can rotate them with the `rotation` keyword to improve readability. In addition to tick labels, axis labels added with `plt.xlabel()` and `plt.ylabel()` give further context to the chart. By customizing ticks and labels, the bar chart becomes much easier to interpret—for instance, showing exactly which drinks correspond to the highest and lowest sales at the coffee shop.

```python
# Plot the bar chart: x-values are indices [0..len(drinks)-1], y-values are sales
plt.bar(range(len(drinks)), sales)

# Create an axes object to customize ticks and labels
ax = plt.subplot()

# Set the x-axis ticks to match the positions of the bars
ax.set_xticks(range(len(drinks)))

# Set the tick labels to the names of the drinks
ax.set_xticklabels(drinks)

# Display the chart with proper labels on the x-axis
plt.show()
```

## Side-By-Side Bars
A bar chart can also be used to compare two datasets across the same categories by placing their bars side by side. This setup is useful when you want to compare values such as populations in different countries over time, prices of the same foods at two restaurants, or enrollments in courses by gender. To achieve this in Matplotlib, you need to calculate separate x-positions for each dataset so that the bars appear next to each other rather than stacked. Each bar has a width (default 0.8), and the positions must be shifted accordingly. For example, if you are plotting two datasets across seven categories, the first set of bars could be at positions `[0, 2, 4, ...]` and the second set shifted by the bar width to `[0.8, 2.8, 4.8, ...]`. This is done using a list comprehension like `[t*element + w*n for element in range(d)]`, where `t` is the number of datasets, `d` is the number of categories, `w` is the bar width, and `n` identifies the dataset (first or second). By adjusting these values, you can easily plot side-by-side comparisons that make differences between datasets clear, such as sales at two branches of a coffee shop.

```python
# Define the categories (drinks) and sales at both locations
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 = [91, 76, 56, 66, 52, 27]   # Sales at store 1
sales2 = [65, 82, 36, 68, 38, 40]   # Sales at store 2

# Generate x-values for the first store's bars
n = 1   # First dataset
t = 2   # Total number of datasets
d = 6   # Number of categories (drinks)
w = 0.8 # Width of each bar
store1_x = [t*element + w*n for element in range(d)]

# Plot store 1's sales bars
plt.bar(store1_x, sales1)

# Generate x-values for the second store's bars
n = 2   # Second dataset
t = 2   # Total number of datasets
d = 6   # Number of categories
w = 0.8 # Width of each bar
store2_x = [t*element + w*n for element in range(d)]

# Plot store 2's sales bars
plt.bar(store2_x, sales2)

# Display the figure with both sets of bars
plt.show()
```

## Stacked Bars
Another way to compare two datasets is by stacking their bars instead of placing them side by side. This approach is useful when you want to show both the breakdown between categories and the total combined value. The technique works by plotting the first dataset as usual, then plotting the second dataset on top with the `bottom` keyword set equal to the heights of the first dataset. For example, if you first plot hours spent on video games, and then plot hours spent reading with `bottom=video_game_hours`, the reading bars begin where the gaming bars end, creating stacked totals. In the MatplotSip case, this means plotting the sales from the first store normally, and then plotting the second store’s sales with `bottom=sales1`, so the bars represent total sales while also showing how each store contributes. This visualization highlights both individual contributions and overall totals in a single chart.

```python

# Plot the first set of bars (Location 1) starting from zero
plt.bar(range(len(sales1)), sales1, label="Location 1")

# Plot the second set of bars (Location 2) stacked on top of Location 1
plt.bar(range(len(sales2)), sales2, bottom=sales1, label="Location 2")

# Add a legend to distinguish the two sets of bars
plt.legend()

# Show the stacked bar chart
plt.show()
```

## Error Bars
Error bars are a way to show the uncertainty or variability in the values represented by a bar chart. They visually indicate how much higher or lower each bar could reasonably be, given measurement imprecision, natural variation, or statistical spread. The vertical line drawn above and below each bar is the error bar, while the small horizontal lines at its ends are called caps, which make the range easier to see.

In Matplotlib, error bars can be added with the `yerr` parameter in `plt.bar()`. If the same uncertainty applies to every bar, `yerr` can be set to a single number, such as `yerr=2`. If different bars have different levels of uncertainty, `yerr` can instead be a list with one value for each bar, matching the order of the bar heights. To make the caps at the ends of the error bars visible, you can also specify `capsize`, which controls their width.

For example, if you plot `[10, 13, 11, 15, 20]` with `yerr=2`, each bar will have an error range of plus or minus two. If you instead set `yerr=[1, 3, 0.5, 2, 4]`, each bar will display a different error size, corresponding to the list values. This ensures that both the central estimate (the bar itself) and the associated uncertainty are clearly communicated.

```python
# Define the categories and milk content per drink
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
ounces_of_milk = [6, 9, 4, 0, 9, 0]   # Amount of milk in ounces for each drink
error = [0.6, 0.9, 0.4, 0, 0.9, 0]    # 10% variation (measurement error)

# Create an axes object to allow custom tick labels
ax = plt.subplot()

# Plot a bar chart:
# - x-values: indices [0..len(drinks)-1]
# - y-values: ounces_of_milk
# - yerr: error values (10% of each bar)
# - capsize=5: adds caps to the error bars
plt.bar(range(len(drinks)), ounces_of_milk, yerr=error, capsize=5)

# Replace x-axis ticks with drink names
ax.set_xticks(range(len(drinks)))
ax.set_xticklabels(drinks)

# Display the bar chart with error bars and labels
plt.show()
```

## Fill between
In addition to bar charts with error bars, line graphs can also display uncertainty using shaded regions. Matplotlib provides the function `plt.fill_between()`, which fills the space between two curves: a lower bound and an upper bound. The function takes three main arguments: the x-values, the list of lower y-values, and the list of upper y-values. Typically, the shaded region represents the possible variation around a central line, such as measurement error or confidence intervals. To make the shading transparent enough to see the line beneath, the `alpha` keyword can be set to a value between 0 and 1. For example, if your data points are `[10, 12, 13, ...]` with an error of ±2, you can calculate the lower bound by subtracting 2 from each value and the upper bound by adding 2. Since lists in Python don’t support direct subtraction or addition with scalars, this is done with list comprehensions, like `[i - 2 for i in y_values]` for the lower bound and `[i + 2 for i in y_values]` for the upper bound. This technique produces a visually appealing way to highlight uncertainty while keeping the trend of the line clear.

```python
# Define the data
months = range(12)
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", 
               "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
revenue = [16000, 14000, 17500, 19500, 21500, 21500, 
           22000, 23000, 20000, 19500, 18000, 16500]

# Create lower and upper bounds with ±10% variation
y_lower = [i - 0.1 * i for i in revenue]  # 10% less
y_upper = [i + 0.1 * i for i in revenue]  # 10% more

# Shade the region of uncertainty between lower and upper bounds
plt.fill_between(months, y_lower, y_upper, alpha=0.2)

# Plot the projected revenue line
plt.plot(months, revenue)

# Create an axes object to customize ticks
ax = plt.subplot()

# Set x-axis ticks and replace them with month names
ax.set_xticks(months)
ax.set_xticklabels(month_names)

# Display the plot
plt.show()
```

## Pie Chart
A pie chart is a way to show how different parts make up a whole by displaying each category as a slice of a circle. It is particularly useful for visualizing proportions, such as demographic distributions, nutritional breakdowns, or survey results. In Matplotlib, pie charts are created with `plt.pie()`, which takes a list of values and automatically converts them into proportions of the circle. For example, providing a list like `[500, 1000, 750, 300, 100]` produces slices in proportion to those amounts. By default, the resulting chart may look stretched or tilted, but this can be corrected by calling `plt.axis('equal')`, which ensures the pie is drawn as a perfect circle rather than an ellipse. This makes the proportions visually accurate and easier to interpret.

```python
# Define the categories and their frequencies
payment_method_names = ["Card Swipe", "Cash", "Apple Pay", "Other"]
payment_method_freqs = [270, 77, 32, 11]

# Create a pie chart using the frequencies
plt.pie(payment_method_freqs)

# Ensure the pie is a circle (not an ellipse)
plt.axis("equal")

# Display the pie chart
plt.show()
```

## Pie Chart Labelling
Pie charts become more meaningful when we clearly indicate what each slice represents. In Matplotlib, there are two main approaches. The first is to create the pie chart and then use `plt.legend()` with a list of category names, which places the labels in a legend alongside the chart. The second is to pass the labels directly into `plt.pie()` using the `labels` argument, which displays them next to the corresponding slices.

Additionally, it’s often helpful to show the percentage that each slice contributes to the whole. This can be done with the `autopct` keyword, which formats the percentages automatically. The format string specifies how the values appear: for example, `'%0.2f'` shows two decimal places without a percent sign, `'%0.2f%%'` shows two decimal places followed by a percent sign, and `'%d%%'` shows rounded whole numbers with a percent sign. By combining `labels` and `autopct` inside `plt.pie()`, we can create a chart that is both descriptive and easy to interpret, showing both the category names and their relative proportions in a single visualization.

```python
# Define the categories and their frequencies
payment_method_names = ["Card Swipe", "Cash", "Apple Pay", "Other"]
payment_method_freqs = [270, 77, 32, 11]

# Create a pie chart with percentage labels (1 decimal place)
plt.pie(payment_method_freqs, autopct='%0.1f%%')

# Keep the pie chart circular
plt.axis("equal")

# Add a legend with the payment method names
plt.legend(payment_method_names)

# Display the chart
plt.show()
```

## Histogram
A histogram is a visualization that shows how values in a dataset are distributed across different ranges. The dataset is divided into equal-sized intervals called bins, and each bin is represented by a bar whose height indicates how many values fall within that range. For example, if your dataset contains numbers between 0 and 100, you could divide it into bins of width 10, with one bar for 0–10, another for 10–20, and so on.

In Matplotlib, histograms are created with `plt.hist()`. By default, it uses 10 bins spanning from the minimum to the maximum of your data. You can customize this with the `bins` argument to control how many intervals to use, and the `range` argument to set the minimum and maximum values included in the plot. For instance, `plt.hist(dataset, range=(66, 69), bins=40)` would zoom in on values between 66 and 69 and use 40 narrow bins for greater detail.

Histograms are especially useful for understanding the overall shape of a dataset—whether values are tightly clustered, spread out evenly, or skewed to one side—helping you spot patterns or areas for deeper analysis.

## Multiple Histograms
When comparing two datasets, plotting multiple histograms on the same axes can highlight differences in their distributions. However, overlapping bars can make the visualization hard to interpret. One way to solve this is by adjusting transparency with the `alpha` parameter: setting `alpha=0.5` makes both histograms partly see-through, so they can be viewed simultaneously. Another option is to use `histtype='step'`, which draws only the outlines of the histograms, preventing them from covering each other.

A second challenge arises when the datasets are very different in size. If one dataset has many more samples, its histogram will naturally dominate, making direct comparison misleading. To address this, you can normalize the histograms with `density=True`, which rescales the bars so that the total area under each histogram equals 1. This way, the shapes of the distributions are directly comparable, regardless of sample size.

Together, these techniques—transparency, step outlines, and normalization—allow for clear and fair comparison of two or more histograms on the same plot, making it easier to identify similarities or differences in the underlying distributions.

## Bar Chart Oerdering
When creating bar charts, the order of the bars can change the way insights are communicated. The approach depends on whether the data is **nominal** or **ordinal**.
For **nominal data**, categories have no natural order, so you have flexibility in how you arrange them. A common strategy is to order them by frequency using `.value_counts()`. For example, if you plot `victory_status` from a chess games dataset and pass `order=df["victory_status"].value_counts(ascending=True).index`, the bars appear in ascending order, making it immediately clear which outcomes are least and most common. By default, `.value_counts()` sorts in descending order, so you can easily flip between the two depending on what you want to highlight.
For **ordinal data**, categories follow a logical sequence that should be preserved. Take student grade levels as an example: “First Year,” “Second Year,” “Third Year,” “Fourth Year.” Although you could sort them by frequency, that would break the natural order. Instead, you pass an explicit list to the `order` parameter in `countplot()` to keep the categorical progression intact.
By aligning your bar chart’s order with the type of data you’re working with, you make the visualization much more informative. Nominal variables benefit from ordering by frequency to highlight dominant categories, while ordinal variables reveal trends and comparisons best when displayed in their natural sequence.

```python
df = pd.read_csv("data_folder/school_data.csv")
print(df.head())
```

## Pitfalls of Pie Charts
Pie charts are widely used to show proportions, but they come with important drawbacks. The human eye is not good at comparing areas or angles, so when a chart has multiple slices of similar size, it becomes difficult to tell which category is larger. This makes pie charts poorly suited for tasks that require precise comparisons.

Bar charts, by contrast, excel at showing relative sizes. Because bars share a common baseline, it is much easier to compare their lengths. For example, if two categories differ only slightly, this is immediately visible on a bar chart but hard to judge in a pie chart.

The key takeaway is that pie charts can work for showing rough proportions or highlighting a single dominant category, but if your audience needs a clear comparison across multiple categories, bar charts are the more effective choice. In practice, this means “taking a pass on the pie” whenever clarity and accuracy in comparing values are more important than simply showing parts of a whole.

When you find yourself in a situation like this, you have a couple of options.
You can aggregate your pie slices to create fewer while still portraying an informative story.
You can see if a bar chart does a more effective job at portraying the information. Bar charts do not suffer nearly as much as pie charts when it comes to visualizing numerous categories. This is because they are not constrained to one circle — instead, numerous bars.

- What proportion of NYC rentals have a gym?
Depending on the type of variable (quantitative or categorical) we want to visualize, we need to use slightly different visualizations.
